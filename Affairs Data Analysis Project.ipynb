```python
import pandas as pd
import plotly.io as pio
pio.renderers.default = 'notebook'
import plotly.express as px
import plotly.graph_objects as go
afs = pd.read_csv('Affairs.csv')
afs
```

```python
# male and female percentage of affairs
afs['have affair'] = afs['affairs'] != 0
MvF = afs[['gender','have affair']].groupby('gender').sum()
numMF =  afs[['gender','have affair']].groupby('gender').count()
perMF = (MvF/numMF)*100
MvF_result = pd.merge(pd.merge(MvF, numMF, on='gender'), perMF.round(1), on = 'gender')
MvF_result.columns = ['# have affair', 'total population', '% have affair']
MvF_result
```

```python
FemData = afs.loc[afs['gender']=='female']
ManData = afs.loc[afs['gender']=='male']
```

```python
def affPerc(*args) :
    if args[0] == 'F' :
        data = FemData
        data1 = f'F{args[1][:3]}1'
        data1 = f'F{args[1][:3]}2'
        data3 = f'F{args[1][:3]}3'
        result = f'F{args[1][:3]}_result'
        print(result)
    else :
        data = ManData
        data1 = f'M{args[1][:3]}1'
        data1 = f'M{args[1][:3]}2'
        data3 = f'M{args[1][:3]}3'
        result = f'M{args[1][:3]}_result'
        print(result)
    data1 = data[[args[1],'have affair']].groupby(args[1]).sum() #total pop
    data2 = data[[args[1],'have affair']].groupby(args[1]).count() ## have affair
    data3 = (data1/data2)*100 # % have affair
    result = pd.merge(pd.merge(data1, data2, on=args[1]), data3.round(1), on = args[1])
    result.columns = ['# have affair', f'total {args[0]} pop', '% have affair']
    return result
```

```python
def figPerc(*args):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=args[0].index, y=args[0]["% have affair"],name='male'))
    fig.add_trace(go.Scatter(x=args[1].index, y=args[1]["% have affair"],name='female'))
    fig.update_layout(
        title=f"{args[2]} vs Affairs",
        xaxis_title=f"{args[2]}",
        yaxis_title="% of Affair",
        legend_title="Gender")
    return fig.show()
```

```python
# Female : occupation and affairs
Focc_result=affPerc('F','occupation')
Focc_result
```

```python
# Male: occupation and affairs
Mocc_result=affPerc('M','occupation')
Mocc_result
```

```python
figPerc(Mocc_result,Focc_result,'Occupation')
```
